```python
import numpy as np
import traitlets
import ipywidgets
import bqplot
```

```python
# how to make a plot in bqplot

# 1: define your data arrays 

x1 = np.arange(100)
y1 = np.random.random(100)*10
y2 = np.random.random(100)*5

# 2: define the scales
x_sc = bqplot.LinearScale()
y_sc = bqplot.LinearScale()

# 3: define the axes
x_ax = bqplot.Axis(label='x', scale=x_sc, tick_format='0.0f')
y_ax = bqplot.Axis(label='y', scale=y_sc, orientation='vertical',
                   tick_format='0.0f')

# 4: define the marks
line = bqplot.Lines(x=x1, y=y1, scales={'x': x_sc, 'y': y_sc}, colors=['pink'])
dots = bqplot.Scatter(x=x1, y=y2, scales={'x': x_sc, 'y': y_sc}, colors=['purple'])

# option: write a function to select an interval of the plot
def on_change_selected(change):
    if change['new'].size < 2: return
    lower = change['new'][0]
    upper = change['new'][1]
    
    label_lower.value = "Lower Limit: %s" % lower
    label_upper.value = "Upper Limit: %s" % upper
    
    selected_points = (x1 > lower) & (x1 < upper)
    dots.selected = np.where(selected_points)[0].tolist()

# define labels using ipywidgets 
label_lower = ipywidgets.Label("Lower Limit:")
label_upper = ipywidgets.Label("Upper Limit:")
    
# interval selector using bqplot
interval_selector = bqplot.interacts.FastIntervalSelector(scale= x_sc)

# option: using the traitlets "observe" function, we can write an interval selector

interval_selector.observe(on_change_selected, ['selected'])
dots.selected_style = {'fill': 'orange'}

# 5: call the figure function
fig = bqplot.Figure(axes=[x_ax, y_ax], marks=[line, dots], title='Random Plot',
                   interaction = interval_selector)


display(ipywidgets.VBox([label_lower, label_upper,
                         ipywidgets.HBox([fig])]))
```
